```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.drop(labels = 'species', axis = 1, inplace = True)
iris.head()
```

```python
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris.head()
```

```python
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
mean = [np.mean(iris['sepal_length']), np.mean(iris['sepal_width']), np.mean(iris['sepal_sum']), np.mean(iris['petal_length']), np.mean(iris['petal_width']), np.mean(iris['petal_sum'])]
# or
mean2 = list(iris.mean()))
summary_dict = {'mean':mean}
#or
summary_df = pd.DataFrame(data = summary_dict)
summary_df
```
